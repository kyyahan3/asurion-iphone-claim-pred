** Note ** : it's been 21 weeks since the iphone 14 was released on October 6, 2022. (As of March 5.)
Our most recent data point is February 13, 2023.
There are 18 weeks between the release of the iphone 14 and our most recent data point, and we are expected to predict 4 more weeks (22 in total.)

Our oldest data point is June 28th, 2021.
Because the iphone 13 and SE were released in 2021, we will have behavior on their first 30 weeks since release. These are the only models that we have for this. If this is the direction our model is headed in, we need to check the amount of data we have for those two ihpones.


Now we need to create a column that has the weeks since the release of the iphone generation. This will be calculated by our newest weeks_monday date, February 13, 2022, minus the release date. The goal of this is for us to be able to see the behavior of claims since the release of the iphone.

```python
#create a new column called weeks_since_release
df['weeks_since_release'] = np.nan

#make the column weeks_since_release a datetime object
df['weeks_since_release'] = pd.to_datetime(df['weeks_since_release'])
```

```python
#This will be calculated by our newest weeks_monday date, February 13, 2022, minus the release date.

#in the column weeks_since_release, subtract the release date from the weeks_monday date
df['weeks_since_release'] = df['weeks_monday'] - df['release']

#make weeks_since_release into the number of weeks and days
df['weeks_since_release'] = df['weeks_since_release'].dt.days / 7
```

```python
#let's look at the apple dataframe
#filter just brand apple
apple = df[df['brand'] == 'apple']
apple
```

```python
#filter the dataset to just weeks_since_release less than 50
apple = apple[apple['weeks_since_release'] < 50]

apple
```
